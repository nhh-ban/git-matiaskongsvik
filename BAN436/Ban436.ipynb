## Function for computing statistics 
### Computes statistics for a chosen DataFrame including: 
- missing values
- percentage missing values
- value count, data types
- descriptive statistics
- optional correlation of numerical variables with a target column.

### The function takes the following parameters as input:
- df (pd.DataFrame): The input DataFrame to analyze.
- target_col (str, optional): The target column for correlation computation. Default is None.
- sort_by: Choose a column to sort in descending order.

### The function returns a DataFrame containing the computed statistics.

```python
## compute_statistics function
def compute_statistics(df, target_col=None, sort_by="NaNs"):

    # create the statistics DataFrame
    stats_df = pd.DataFrame(df.isnull().sum(), columns=["NaNs"])
    stats_df["percent_NaN"] = round(df.isnull().sum() / df.shape[0] * 100, 2)
    stats_df["value_count"] = df.count()
    stats_df["datatype"] = df.dtypes
    stats_df["mean"] = round(df.mean(numeric_only=True), 1)
    stats_df["min"] = df.min(numeric_only=True)
    stats_df["max"] = df.max(numeric_only=True)
    
    # compute correlation with the target column if provided
    if target_col and target_col in df.columns:
        correlations = df.corr(numeric_only=True)[target_col]
        stats_df["Corr_" + target_col] = correlations
    
    x = str(sort_by)
    
    # sort by missing values in descending order
    stats_df = stats_df.sort_values(by=x, ascending=False)

    return stats_df
```

## Function for plot_top_correlated_histograms. 
### The function takes the following parameters as input:
- df (pd.DataFrame): The input DataFrame to plot.
- target_col (str, optional): The target column for correlation computation. Default is None.
- top_n: Number of feature variables to include.
### Returns histogram plots.

```python
def plot_top_correlated_histograms(df, target_col, top_n=20):

    # compute correlation of numerical columns with the target column
    correlations = df.corr(numeric_only=True)[target_col].sort_values(ascending=False)
    
    # select top correlated features (excluding the target itself)
    top_correlated_cols = correlations.index[1:top_n + 1]
    
    # create subplots
    num_cols = len(top_correlated_cols)

    # determine the number of rows for subplots
    rows = (num_cols // 5) + (1 if num_cols % 5 != 0 else 0)  
    fig, axes = plt.subplots(nrows=rows, ncols=5, figsize=(20, 5 * rows))
    axes = axes.flatten()
    
    # plot histograms
    for ax, col in zip(axes, top_correlated_cols):
        ax.hist(df[col], bins=20, color="blue", alpha=0.7, edgecolor="black")
        ax.set_title(col)
        ax.set_xlabel("Value")
        ax.set_ylabel("Frequency")

    plt.tight_layout()
    plt.show()
```

### Function for plot_categorical_value_counts.

### Function to predict values of `SalePrice`.
The predict_sale_price function is designed to evaluate the predictive power of an individual explanatory variable in a DataFrame, in this case the dependent target variable `SalePrice`, using a simple linear regression model. The linear regression model is formulated as:

$$
SalePrice_i = \beta_0 + \beta_1 * explanatoryVariable_i + \epsilon_i
$$

Where:

- \( SalePrice_i \): The dependent variable (predicted value) for observation \( i \)
- \( \beta_0 \): The intercept (value of \( y \) when \( x = 0 \))
- \( \beta_1 \): The slope of the line (change in \( y \) for a unit change in \( x \))
- \( x_i \): The independent variable (predictor) for observation \( i \)
- \( \epsilon_i \): The error term (difference between the actual and predicted values)

In addition to evaluating the RMSE for each linear regression model, as specified in the task description, I've also dded r**2 score to indicate the explanatory power of each variable on the prediction of the dependent target variable `SalePrice`.


### The function performs the following key tasks:

### Iterative Modeling:

- For each explanatory variable in the dataset (excluding SalePrice), the function fits a simple linear regression model.

### Model Evaluation:

- It splits the data into training and test sets.
- Trains a linear regression model for each variable individually.

- Calculates model performance metrics and other relevant information including:
    - `RMSE`: Measures the average difference between predicted and actual values.
    - `R²` Score: Indicates the proportion of variance in SalePrice explained by the variable.
    - `Intercept`, of model
    - `Coefficient` of model
- Evaluates and track each model's performance using Root Mean Square Error (RMSE).
- Identifies the variable with the best performance (lowest RMSE).

### Outputs
- `best_model_info` containing information about the best perfoming variable stored in a dictionary, including the `variable name`, `RMSE`, `R²` score, `intercept`, `coefficient`, model used, test data, and prediction values.
- `all_results`, a list of dictionaries sorted by the RMSE of each variable, each containing all the results from from their respective models.

These outputs are used as parameters in the functions named show_best_model_info and show_results which displays the model results and linear regression equation of the best performing variable and the model results for all other variables sorted by `RMSE` in ascending order.

### Visualization:

Two separate function takes the best_model_info returned by predict_sale_price to plot the predicted vs. actual values, and the regression line for the best-performing model to visually assess its accuracy.

```python
def predict_sale_price(df):
    # define target variable
    target_name = "SalePrice"

    # separate DataFrame for values of the dependent variable 
    y = df[target_name]

    # set seed to ensure reproducibility
    seed = 2
    
    # split y once into training and testing sets (20/80 split)
    y_train, y_test = train_test_split(y, test_size=0.2, random_state=seed)
    
    # get list of columns for all explanatory variables
    explanatory_variables_cols = [col for col in df.columns if col != target_name]

    # empty list to store results
    results = []
    best_model_info = {
                "variable": None, 
                "rmse": float("inf"),
                "r2": float("-inf"), 
                "intercept": None, 
                "coefficient": None, 
                "model": None, 
                "X_test": None, 
                "y_test": None, 
                "y_pred": None
            }

    # Loop through each explanatory variable
    for variable in explanatory_variables_cols:
        X = df[[variable]]

        # Split X into training and testing sets (consistent split for X and y)
        X_train, X_test = train_test_split(X, test_size=0.2, random_state=seed)

        # train the regression model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # predict y-values using the trained model
        y_pred = model.predict(X_test)

        # Calculate metrics
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        coefficient = round(model.coef_[0], 2)
        intercept = round(model.intercept_,2)
        results.append(
            {
                "variable": variable, 
                "rmse": rmse, 
                "r2": r2, 
                "intercept": intercept, 
                "coefficient": coefficient, 
                "model": model,
                "X_test": X_test,
                "y_test": y_test,
                "y_pred": y_pred
            }
        )

        # track the best model and update best_model_info dict
        if rmse < best_model_info["rmse"]:
            best_model_info.update({
                "variable": variable,
                "rmse": rmse,
                "r2": r2,
                "intercept": intercept, 
                "coefficient": coefficient,
                "model": model,
                "X_test": X_test,
                "y_test": y_test,
                "y_pred": y_pred
            }
        )
    ## sort the result list of dictionaries by the key rmse in ascending order
    all_results = sorted(results, key = lambda x: x["rmse"])

    return best_model_info, all_results
```

```python
def show_best_model_info(best_model_info):
    # best variable and its metrics
    print(f"\nBest Variable: {best_model_info["variable"]}, RMSE: {best_model_info["rmse"]:.2f}, R^2: {best_model_info["r2"]:.2f}")
    print("\n")
    # print the estimated regression equation using values in the best_model_info dictionary
    print(f"Estimated Regression Equation for {best_model_info["variable"]}:")
    print("")
    print(f"{target_name} = {best_model_info["intercept"]} + {best_model_info["coefficient"]} × {best_model_info["variable"]}")
    print("\n")
```

```python
def show_results(all_results):
    ## iterate over the list ad print the variable name, RMSE, and R^2 values
    for res in all_results:
        print(f"Variable: {res["variable"]}, RMSE: {res["rmse"]:.2f}, R^2: {res["r2"]:.2f}")
```

```python
def in_sample_prediction_plot(best_model_info):

    # plot in-sample predictions for the best model
    plt.figure(figsize=(8, 8))
    plt.scatter(best_model_info["y_pred"], best_model_info["y_test"], alpha=0.7, edgecolor="k", label="Predicted vs Actual")
    plt.plot(best_model_info["y_test"], best_model_info["y_test"], color="red", linewidth=2, label="45-degree line (perfect prediction)")
    plt.title(f"In-sample Predictions for {best_model_info["variable"]}", fontsize=16)
    plt.ylabel("Actual SalePrice", fontsize=14)
    plt.xlabel("Predicted SalePrice", fontsize=14)
    plt.legend()
    plt.grid(alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
def regression_line_best_model(best_model_info):

    # plot the regression line for the best model
    X_range = np.linspace(df[best_model_info['variable']].min(), df[best_model_info['variable']].max(), 100)
    y_range = best_model_info['intercept'] + best_model_info['coefficient'] * X_range
    plt.figure(figsize=(8, 8))
    plt.scatter(df[best_model_info['variable']], df[target_name], alpha=0.7, label="Data Points")
    plt.plot(X_range, y_range, color="purple", linewidth=2, label=f"Estimated Regression Line: {target_name} = {best_model_info['intercept']} + {best_model_info['coefficient']} × {best_model_info['variable']}")
    plt.title(f"Regression Line for {best_model_info['variable']}", fontsize=16)
    plt.xlabel(best_model_info['variable'], fontsize=14)
    plt.ylabel("SalePrice", fontsize=14)
    plt.legend()
    plt.grid(alpha=0.6)
    plt.tight_layout()
    plt.show()
```
